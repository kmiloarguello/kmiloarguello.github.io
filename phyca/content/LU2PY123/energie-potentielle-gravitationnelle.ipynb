```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constants
G = 6.67430e-11  # m^3 kg^-1 s^-2

# Masses

# Earth
M = 5.972e24  # kg

# Moon
m = 7.342e22  # kg
```

```python
def E(r):
  return -G * M * m / r
```

```python
x = np.linspace(1e6, 1e8, 1000)
y = E(x)

plt.plot(x, y)
plt.xlabel('r (m)')
plt.ylabel('E (J)')
plt.title('Gravitational potential energy between Earth and Moon')
plt.show()
```
