```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Définition des paramètres
m = 1.0  # masse
L = 1.0  # moment cinétique réduit
phi = 1.0  # vitesse angulaire
w_0 = 1.0  # fréquence angulaire
```

```python
# Création d'une plage d'angles theta
theta = np.linspace(0, 2*np.pi, 100)

# Calcul de V_eff(theta) pour chaque valeur d'angle theta
V_eff = ((m * phi * L) / 2) + (w_0**2 * np.cos(theta))

# Tracé du graphique
plt.plot(theta, V_eff)
plt.xlabel('$\Theta$')
plt.ylabel('$U_{eff} \ (\Theta) $')
plt.title('Energie Potentielle Effective en fonction de l\'angle $\Theta$')
plt.grid(True)
plt.show()
```
