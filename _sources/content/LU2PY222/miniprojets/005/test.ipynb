```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Tao = .2 # s
```

```python
# 1.1
def g(t):
    return ( 1 / Tao ) * np.exp( -t / Tao )

def h(x):
    return - Tao * np.log(1 - x)
```

Avec une liste de `1000` valeurs dépuis `0` jusqu'à `1`

```python
# 1.1
t = np.arange(0,1,0.0001)

_g = g(t)
_h = h(t)

plt.plot(t,_g, label='g(t)')
plt.plot(t,_h, label='h(t)')
plt.xlabel('Temps (s)')
plt.legend()
plt.show()
```

```python
# 1.2 et 1.3
def get_random_list(n=10000):
    x = np.zeros(n)
    for i in range(n):
        x[i] = np.random.uniform(0,1)
    return x

x = get_random_list()
y = h(x)

plt.hist(y, bins=len(x))
plt.show()
```

```python
# 1.4
moyenne = np.mean(y) 

# Pour tracer la moyenne
y_mean = np.arange(10) # pour ligne droit verticale
x_mean = np.empty(y_mean.shape)
x_mean.fill(moyenne)

plt.hist(y, bins=len(x))
plt.plot(x_mean,y_mean, 'red', label='moyenne')
plt.legend()
plt.show()
```

La ligne rouge se trouve dans la valeu attendue de la distribution
