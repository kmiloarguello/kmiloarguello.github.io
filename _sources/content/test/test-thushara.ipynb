```python
# Création d'un tableau - DataFrame
from scipy.optimize import curve_fit
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate sample data for a straight line (y = mx + c)
m = 2  
c = 5  
x_data = np.linspace(0, 10, 40)  # Generate x values
y_data = - m * x_data + c  # Calculate corresponding y values

df = pd.DataFrame(np.random.randn(100, 2), columns=list('XY'))
df

# Add some noise to the data
np.random.seed(0)
Y =  np.random.normal(size=x_data.size)
y_data_noisy = y_data + Y

# Define the model function for a straight line (to be used for Gaussian fitting)
def model_func(x, m, c):
    return m * x + c

# Perform Gaussian fitting using curve_fit
params, _ = curve_fit(model_func, x_data, y_data_noisy)

# Extract the fitted parameters
fitted_m, fitted_c = params

# Plot the original data and the Gaussian fit
plt.scatter(x_data, y_data_noisy, color='r', label='Noisy Data')
plt.plot(x_data, model_func(x_data, fitted_m, fitted_c), label='Fitted Line', color='b')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gaussian Fitting of a Straight Line')
plt.legend()
plt.grid(True)
plt.show()

print('Parameters of the Gaussian Fit - m:', fitted_m, 'c:', fitted_c)
```

```python
data_avec_dataframe = pd.DataFrame({'x': x_data, 'y': y_data_noisy})

data_avec_dataframe.head()

# Modèle
params2, _ = curve_fit(model_func, data_avec_dataframe['x'], data_avec_dataframe['y'])
fitted_m2, fitted_c2 = params2

print('Parameters of the Gaussian Fit - m:', fitted_m2, 'c:', fitted_c2)

# plot
data_avec_dataframe.plot(x='x', y='y', kind='scatter', color='r')
plt.plot(data_avec_dataframe['x'], model_func(data_avec_dataframe['x'], fitted_m2, fitted_c2), label='Fitted Line', color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gaussian Fitting of a Straight Line')
plt.grid(True)
plt.show()
```
